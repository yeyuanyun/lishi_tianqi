```python
import pandas as pd
```

```python
df = pd.read_csv('南昌市历史天气.csv')
df
```

```python
# 爬取时名字写错了，修改过来
df['lowest_temp'] = df['lowset_temp']
df.drop('lowset_temp', axis=1, inplace=True)
```

```python
# print(df.isnull().any())
# 查看该列所有取值
print(df['wind_direction'].unique())
print(df['air_quality'].unique())
```

```python
# 将日期与星期分离
def split_date_weekday(row):
    date, weekday = str(row['weather_date']).split(" ")
    return date, weekday

# 删去温度的单位
def standard_temp(row):
    high_temp = int(row['highest_temp'].replace('°',''))
    low_temp = int(row['lowest_temp'].replace('°',''))
    return high_temp, low_temp

# 分离天气
def standard_weeather(weather):
    return weather.split('~')

# 分离风向 风力
def split_wind_direction_power(row):
    if row['wind_direction'].startswith('无持续风向'):
        direction = '无持续风向'
        power = row['wind_direction'][5:]
    else:
        direction, power = row['wind_direction'].split('风', 1)
        direction = direction + '风'
        
    return direction, power

def split_air_quality(row):
    if row['air_quality'] == '-':
        return (0,'')
    air_measure, air_degree = row['air_quality'].split(' ')
    return air_measure, air_degree

# 用历年数据填补空气指数为零的项
def fill_null_air_measure_degree(row):
    if row['air_measure'] == 0:
        same_day_df = df[df['month_day'] == row['month_day']]
        non_zero_values = same_day_df['air_measure'][same_day_df['air_measure'] != 0]
        row['air_measure'] = int(non_zero_values.mean())
        # 通过比对表格发现50及以下是优，100及以下是良，150及以下是轻度，发现这是按照美国环境保护署（EPA） 制定的 24 小时 PM2.5 浓度标准。
        # 当空气质量指数在 0 至 50 的范围内时，就被称为 "良好"；51 至 100 为 "中等"；101 至 150 为 "对敏感人群不健康"；151 至 200 为 "不健康"；201 至 300 为 "非常不健康"；301 以上为 "有害"。
        if row['air_measure'] <= 50:
            row['air_degree'] = '优'
        elif row['air_measure'] <= 100:
            row['air_degree'] = '良'
        elif row['air_measure'] <= 150:
            row['air_degree'] = '轻度'
        elif row['air_measure'] <= 200:
            row['air_degree'] = '中度'
        elif row['air_measure'] <= 300:
            row['air_degree'] = '重度'
        else:
            row['air_degree'] = '有害'
    
        row['air_quality'] = str(row['air_measure']) + ' ' + str(row['air_degree'])
    return row
```

```python
df[['highest_temp', 'lowest_temp']] = df.apply(standard_temp, axis=1, result_type='expand')
df['temp_difference'] = df['highest_temp'] - df['lowest_temp']
df['average_temp'] = (df['highest_temp'] + df['lowest_temp']) / 2
```

```python
df[['date', 'weekday']] = df.apply(split_date_weekday, axis=1, result_type='expand')
df['date'] = pd.to_datetime(df['date'])
df[['direction', 'power']] = df.apply(split_wind_direction_power, axis=1, result_type='expand')
df[['air_measure', 'air_degree']] = df.apply(split_air_quality, axis=1, result_type='expand')
df['month_day'] = df['date'].dt.strftime('%m-%d')
df['year'] = df['date'].dt.strftime('%Y')
df['year_month'] = df['date'].dt.strftime('%Y-%m')
df['date'] = df['date'].dt.date
df['air_measure'] = df['air_measure'].astype(int) 
```

```python
# 查看没有空气质量指数的列
# print(df.loc[df["air_measure"] == ''])
# 可知有两年的12月份没有空气质量指数，采用同时期的平均值填补
```

```python
df = df.apply(fill_null_air_measure_degree, axis=1)
```

```python
df.dtypes
```

```python
# 修改列名顺序
# 不需要'weather_date'了，直接从列名中去除
new_order = ['date', 'year_month', 'month_day', 'weekday', 'weather', 'highest_temp', 'lowest_temp', 'average_temp', 'temp_difference','wind_direction', 'power', 'direction', 'air_quality', 'air_measure', 'air_degree', 'year']

# 使用 reindex 方法重新排列列名顺序
df = df.reindex(columns=new_order)
df.head()
```

```python
df.to_csv('南昌市历史天气_clean.csv', index=False,  encoding='utf-8')
```
